```python
data1 = data.iloc[:,1:588]
print(data1.columns)
## Ver nombre de columnas (solo las que tienen fecha)
```

```python
data1.columns = pd.to_datetime(data1.columns).strftime("%Y/%m/%d")
data1.head(5)
```

```python
data2 = data1.groupby(pd.PeriodIndex(data1.columns, freq="d"), axis=1).sum()
data2.head(5)
```

```python
sum_monthly = data2.sum(axis=0)
print(sum_monthly)
## sumo lo que hay dentro por mes
```

```python
sort_monthly = sum_monthly.sort_values(ascending=False)/2
sort_monthly
## suma por dia
```
